```python
import os

import pysd
import pysd_package.vis as vis
import pickle

import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from IPython import display
```

```python
with open("experiment.out", "rb") as fp:
    exp_afm = pickle.load(fp)
```

```python
fig = plt.figure(figsize=(5, 4), dpi=400)
hs = [x.hx for x in exp_afm['configs']]
ms = [x.avg_M()[0] for x in exp_afm['restartfiles']]
# plt.title("Hysteresis loop for various D and T={} K, deltaJ={}".format(T, deltaJ))
plt.xlabel(r"$H_x$")
plt.ylabel(r"$\mathrm{M_x}$")

ms_fm = [1 if h >= 0 else -1 for h in hs]
plt.plot(np.array(hs)/235.0314, ms, label="AFM -> FM")

plt.plot(np.array(hs)/235.0314, ms_fm, label="FM", color="red", linestyle="dotted")

plt.legend()

```
