```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def parse_Mz(filename):
    res = []
    data = np.genfromtxt(filename)
    ens = np.unique(data[:,1])
    for e in ens:
        res.append(np.average(data[data[:,1] == e,6]))
    return res

def parse_hz(filename):
    data = np.genfromtxt(filename)
    return data

def plot_hyst(rundir, label=""):
    Mzs = []
    Hzs = []
#     rundir = "run_dm_long"
    for i in range(0, 10000):
        try:
            Mzs.append(parse_Mz(rundir+"/"+str(i)+"/"+"/restart.EuAlOBas.out"))
            Hzs.append(parse_hz(rundir+"/"+str(i)+"/"+"/hz"))
        except:
            break
    plt.plot(Hzs, Mzs, label=label)
    return np.array(Hzs), np.array(Mzs)
    
```

```python
DMs = ["0", "0.002", "0.005", "0.02", "0.04", "0.07", "0.1", "0.4", "0.5", "0.6", "0.63", "0.64", "0.65"]
Mzs = {}
Hzs = []
for dm in DMs:
    plt.figure(figsize=(15, 7))
    
    Hzs, Mzs[dm] = plot_hyst("runs/" + dm + "/run")
    plt.xlim(-0.015, 0.015)
    plt.vlines(0, -1, 1, color="black")
    plt.hlines(0, -1, 1, color="black")

    plt.xlabel("H [T]")
    plt.ylabel("M")
    plt.title("D="+dm)

    plt.savefig("hyst_" + dm + ".png")
    
```
