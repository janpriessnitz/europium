```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def parse_Mz(filename):
    res = []
    data = np.genfromtxt(filename)
    ens = np.unique(data[:,1])
    for e in ens:
        res.append(np.average(data[data[:,1] == e,6]))
    return res

def parse_hz(filename):
    data = np.genfromtxt(filename)
    return data
```

```python
Mzs = []
Hzs = []
rundir = "run_dm_long"
for i in range(0, 10000):
    try:
        Mzs.append(parse_Mz(rundir+"/"+str(i)+"/"+"/restart.EuAlOBas.out"))
        Hzs.append(parse_hz(rundir+"/"+str(i)+"/"+"/hz"))
    except:
        break
```

```python
Mzs2 = []
Hzs2 = []
rundir = "run_nodm"
for i in range(0, 10000):
    try:
        Mzs2.append(parse_Mz(rundir+"/"+str(i)+"/"+"/restart.EuAlOBas.out"))
        Hzs2.append(parse_hz(rundir+"/"+str(i)+"/"+"/hz"))
    except:
        break
```

```python
Mzs3 = []
Hzs3 = []
rundir = "run_nodm_long"
for i in range(0, 10000):
    try:
        Mzs3.append(parse_Mz(rundir+"/"+str(i)+"/"+"/restart.EuAlOBas.out"))
        Hzs3.append(parse_hz(rundir+"/"+str(i)+"/"+"/hz"))
    except:
        break
```

```python
plt.scatter(Hzs, Mzs)
plt.plot(Hzs2, Mzs2)
plt.plot(Hzs3, Mzs3)
plt.xlim(0, 0.025)
```
