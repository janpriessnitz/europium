```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
coordfile = "../runs/basic/coord.EuAlOBas.out"
structfile = "../runs/basic/struct.EuAlOBas.out"
dmdatafile = "../runs/basic/dmdata.EuAlOBas.out"
```

```python
coord = np.genfromtxt(coordfile)
struct = np.genfromtxt(structfile)
dmdata = np.genfromtxt(dmdatafile)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(coord[:,1], coord[:,2], color="black")

for exc in range(len(struct)):
    i, j = struct[exc][0:2]
    x0, y0 = coord[int(i)-1][1:3]
    x1, y1 = coord[int(j)-1][1:3]
    plt.plot([x0, x1], [y0, y1], color="red", linewidth=1)
    
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(coord[:,1], coord[:,2], color="black")

for exc in range(len(dmdata)):
    i, j = dmdata[exc][0:2]
    dx, dy, dz = dmdata[exc][7:10]
    x0, y0 = coord[int(i)-1][1:3]
    x1, y1 = coord[int(j)-1][1:3]
    plt.plot([x0, x1], [y0, y1], color=("red" if dz > 0 else "blue"), linewidth=1, alpha=0.5)
    
plt.show()
```
