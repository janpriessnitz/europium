```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def parse_Mz(filename):
    res = []
    data = np.genfromtxt(filename)
    ens = np.unique(data[:,1])
    for e in ens:
        res.append(np.average(data[data[:,1] == e,6]))
    return res
```

```python
Mzmap = {}
rundir = "runs"
for dm in os.listdir(rundir):
    for hz in os.listdir(rundir+"/"+dm+"/"):
        try:
            Mzmap[(float(dm), float(hz))] = parse_Mz(rundir+"/"+dm+"/"+hz+"/restart.EuAlOBas.out")
        except:
            pass
```

```python
Mzmap.keys()
```

```python
dm = 0.1
```

```python

def plot_mag(dm):
    xs = []
    ys = []
    for k, v in Mzmap:
        if k != dm:
            continue
        xs.extend([v]*len(Mzmap[(k, v)]))
        ys.extend(Mzmap[(k, v)])
    plt.scatter(xs, ys, label=("D = " + str(dm) + " mRy"))
```
